```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
data=pd.read_csv('/content/movies.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
features = [
    'genres',               # High weight for genres
    'keywords',             # Moderate weight for keywords
    'overview',             # Moderate weight for overview
    'tagline',              # Low weight for tagline
    'cast',                 # High weight for cast
    'director',             # Low weight for director
    'production_companies',  # Moderate weight for production companies
    'production_countries',  # Moderate weight for production countries
    'runtime',              # Low weight for runtime
    'original_language',    # Low weight for original language
    'vote_average'          # Low weight for average rating
]
```

```python
for f in features:
  data[f]=data[f].fillna('')
```

```python
data['weighted_features'] = (
    data['genres'].fillna('').astype(str) * 5 + ' ' +  # High weight for genres
    data['keywords'].fillna('').astype(str) * 4 + ' ' + # Moderate weight for keywords
    data['overview'].fillna('').astype(str) * 4 + ' ' +  # Moderate weight for overview
    data['tagline'].fillna('') + ' ' +                   # Low weight for tagline
    data['cast'].fillna('').astype(str) * 4 + ' ' +     # High weight for cast
    data['director'].fillna('') + ' ' +                  # Low weight for director
    data['production_companies'].fillna('').astype(str) * 2 + ' ' +  # Moderate weight for production companies
    data['production_countries'].fillna('').astype(str) * 2 + ' ' +  # Moderate weight for production countries
    data['runtime'].fillna(0).astype(str) + ' ' +        # Low weight for runtime
    data['original_language'].fillna('').astype(str) + ' ' +  # Low weight for original language
    data['vote_average'].fillna(0).astype(str) + ' '      # Low weight for average rating
)

```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['weighted_features'])
```

```python
similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
print(similarity)
```

```python
ip=input('Enter movie name: ')
list_of_all_movies=data['title'].tolist()
close_match=difflib.get_close_matches(ip,list_of_all_movies)
final_input=close_match[0]
idx=data[data['title']==final_input].index.values[0]
score_list=list(enumerate(similarity[idx]))
sorted_list=sorted(score_list,key=lambda x:x[1],reverse=True)
print('MOVIES FOR YOU ARE - \n')
i=1
for m in sorted_list:
  index=m[0]
  fav=data[data.index==index]['title'].values[0]
  if(i<21):
    print(i,'.',fav)
    i+=1
```
